```python
'''
Por Ejemplo, una emnpresa priada fabricante de vacunas contra el 
covid-19 desea saber cuantas personas de las que se aplicaron dicha
vacunas poseen efectos secundarios y cuales. Entonces se procede a
tomar una muestra de los pacientes vacunados para determinar esto a 
traves de una encuesta.
Sabemos que hoy en dia es mucho mas facil realizar encuestas a nivel 
mas grande, pero que sirva de ejemplo.  
'''

#Esta seria la poblacion vacunada, al ser un volumne muy grande
#se procede a muestrear de forma descriptiva 

poblacionVacunada = range(1,10000)

for valor in poblacionVacunada:
    print('Ciudadano N°', valor)
```

```python
'''
Vamos a generar de forma aleatoria la encuesta que se llevaria a cabo
Tomamos n Ciudadanos y n Sintomas
'''

import random

sintomas=('Leve Dolor de Cabeza','Dolor de Estómago','Dolor Muscular','Cansancio','Sin Efectos Secundarios')

sintomasMuestreo=[]

#Muestra de n pacientes vacunados
for i in range(100):
    
    #Generamos los valores random y anexamos a una lista que utilizaremos luego
    sintomaAleatorio=random.choice(sintomas)
    
    sintomasMuestreo.append(sintomaAleatorio)
    
    print('\nCiudadano Vacunado N°',random.choice(poblacionVacunada)
         ,'-->Sintoma Asociado:',sintomaAleatorio)
    
    
    
```

```python
'''
Contabilizamos la cantidad de Sintomas por Vacunados segun la muestra descriptiva

'''

print('\n====================================================================================')

print('-Cantidad Total de Vacunados de la Muestra con Leve Dolor de Cabeza:'
     ,sintomasMuestreo.count('Leve Dolor de Cabeza')
     ,'('
     ,((sintomasMuestreo.count('Leve Dolor de Cabeza')) * len(sintomasMuestreo)) / 100
     ,'%)'
     )


print('-Cantidad Total de Vacunados de la Muestra con Dolor de Estómago:'
     ,sintomasMuestreo.count('Dolor de Estómago')
     ,'('
     ,((sintomasMuestreo.count('Dolor de Estómago')) * len(sintomasMuestreo)) / 100
     ,'%)'
     )


print('-Cantidad Total de Vacunados de la Muestra con Dolor Muscular:'
     ,sintomasMuestreo.count('Dolor Muscular')
     ,'('
     ,((sintomasMuestreo.count('Dolor Muscular')) * len(sintomasMuestreo)) / 100
     ,'%)'
     )


print('-Cantidad Total de Vacunados de la Muestra con Cansancio:'
     ,sintomasMuestreo.count('Cansancio')
     ,'('
     ,((sintomasMuestreo.count('Cansancio')) * len(sintomasMuestreo)) / 100
     ,'%)'
     )



print('-Cantidad Total de Vacunados de la Muestra Sin Efectos Secundarios:'
     ,sintomasMuestreo.count('Sin Efectos Secundarios')
     ,'('
     ,((sintomasMuestreo.count('Sin Efectos Secundarios')) * len(sintomasMuestreo)) / 100
     ,'%)'
     )

'''
Aca generamos un set que es el conjunto de los elementos únicos de la lista sintomas, y aplicamos
un max() usando un método de toda lista el count(elemento) que nos devuelve la cantidad de elementos 
en la lista.
'''
print('\n==============================================================================')
print('-Sintoma más presentado en dicha Muestra:'
      , max(set(sintomasMuestreo),key=sintomasMuestreo.count))
print('==============================================================================')
```

```python
'''
Generamos una gráfica de Barras 
'''
import matplotlib.pyplot as plt 

x = sintomas

y = [
    sintomasMuestreo.count('Leve Dolor de Cabeza')
    ,sintomasMuestreo.count('Dolor de Estómago')
    ,sintomasMuestreo.count('Dolor Muscular')
    ,sintomasMuestreo.count('Cansancio')
    ,sintomasMuestreo.count('Sin Efectos Secundarios')
    
]
plt.title('Sintomas Asociados en grupos de Vacunados(Muestra Descriptiva)',size=18)

plt.ylabel('Sintomas Asociados',size=12)
plt.xlabel('Cantidad de Vacunados[Por Agrupaciones]',size=12)

plt.barh( x , y , color='red')
```
